#### Atributos

A la hora de añadir atributos a la clase, es importante distinguir dos tipos:

* __Atributos de instancia__: Pertenecen a la instancia de la clase o al objeto. Son atributos particulares de cada instancia, en nuestro caso de cada persona.

* __Atributos de clase__: Se trata de atributos que pertenecen a la clase, por lo tanto serán comunes para todos los objetos.

```python
# atributos de instancia

class Persona:

  def __init__(self, nombre, apellido, dni):
    self.nombre = nombre
    self.apellido = apellido
    self.dni = dni
```

Herencia

__¿Qué es?__

La herencia es un proceso mediante el cual se puede crear una clase hija que hereda de una clase padre, compartiendo sus métodos y atributos.

Además de ello, una clase hija puede sobreescribir los métodos o atributos, o incluso definir unos nuevos.

```python
class CalculadoraBase:

  def sumar(self, numero1, numero2):
    return numero1 + numero2

  def restar(self, numero1, numero2):
    return numero1 - numero2
```

```python
calc = CalculadoraBase()
```

```python
calc.sumar(5,5)
```

```python
calc.restar(10,5)
```

```python
class CalculadoraAvanzada:

  def sumar(self, numero1, numero2):
    return numero1 + numero2

  def restar(self, numero1, numero2):
    return numero1 - numero2

  def multiplicar(self, numero1, numero2):
    return numero1 * numero2

  def dividir(self, numero1, numero2):
    return numero1 / numero2

```

```python
calc_avanzada = CalculadoraAvanzada()
```

```python
calc_avanzada.sumar(6,6)
```

```python
calc_avanzada.multiplicar(2,6)
```

la clase CalculadoraAvanzada repite métodos de CalculadoraBase

Sobreescribir un método

```python
class Persona:

  def __init__(self, nombre, dni):
    self.nombre = nombre
    self.dni = dni
    self.clave = f"{nombre} -- {dni}"

  def ver_persona(self):
    print(f"Nombre: {self.nombre}, DNI: {self.dni}")
```

```python
class Alumno(Persona):

  def __init__(self, nombre, dni, curso):
    self.curso = curso
    super().__init__(nombre, dni)

  def ver_curso(self):
    print(f"el curso es {self.curso}")

  def ver_persona_alumno(self):
    super().ver_persona()
    print(f"nombre: {self.nombre} -- curso: {self.curso}")

  def ver_persona(self):
    print(f"Nombre: {self.nombre}, DNI: {self.dni} -- curso: {self.curso}")
```

```python
alumno = Alumno("valentin", 35435, "python avanzado")
```

```python
alumno.ver_persona()
```

Herencia Múltiple

```python
class Clase_A:

  def __init__(self, a):
    self.a = a

  def metodo_A(self):
    print("soy el metodo A")
```

```python
class Clase_B:

  def __init__(self, b):
    self.b = b

  def metodo_B(self):
    print("soy el metodo b")
```

```python
class Clase_C(Clase_A, Clase_B):

  def __init__(self, a, b, c):
    self.c = c
    Clase_A.__init__(self, a)
    Clase_B.__init__(self, b)

  def metodo_C(self):
    print("soy el metodo c")
```

```python
foo = Clase_C("a", "b", "c")
```

```python
foo.a
```

```python
foo.c
```

```python
foo.b
```

```python
foo.metodo_C()
```

```python
foo.metodo_B()
```

```python
foo.metodo_A()
```

Callables

👉 Al definir __call__, hacés que el objeto se comporte como una función, que esta inicializada

```python
class Multiplicador:

  def __init__(self, factor):
    self.factor = factor

  def __call__(self, valor):
    return valor * self.factor

  def multiplicar(self, valor):
    return valor * self.factor
```

```python
doble = Multiplicador(2)
```

```python
doble(5)
```

```python
triple = Multiplicador(3)
```

```python
triple(5)
```

```python
mult = Multiplicador(4)
```

```python
mult.multiplicar(4)
```
